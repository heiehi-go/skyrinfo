# 计算斯格明子半径

导入计算斯格明子半径的函数

```python
from skyrinfo import cal_r,outline,outline_filter,outline_sorted,cal_center
import matplotlib.pyplot as plt
```

导入数据

```python
import micromagneticdata as md

data = md.Data(name="A_K",dirname="C:\\MagneticData\\EX_UNI")
```

```python
x,y= outline_sorted(data[0][0])
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
data[0][0].sel(z=0).z.mpl(ax=ax)
ax.plot(x*1e9,y*1e9)
```

计算

```python
list_r = []
for drive in data:
    r = cal_r(drive[0])

    print(r)
    list_r.append(r)
    
```

```python
cal_r(drive[0].sel(z=0).z)
```

```python
list_r
```

```python
import numpy as np

list_r = np.array(list_r)
list_r = np.reshape(list_r,[5,5])
```

画出半径的变化图

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.imshow(list_r,interpolation='gaussian')
```
